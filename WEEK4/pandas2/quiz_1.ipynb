```python
number = 10
names = ['Adam', 'Alan', 'Alex', 'Alvin', 'Andrew', 'Anthony', 'Arnold', 'Baldy', 
         'Baron', 'Billy', 'Boris', 'Bruno', 'Caley', 'Champ', 'Charlie', 'Clark']

df = pd.DataFrame(columns = ['Name', 'Math', 'Science', 'English', 'Total'])

for _ in range(number):
    
    name = np.random.choice(names)
    while name in list(df['Name']):
        name = np.random.choice(names)
    
    df.loc[len(df)] = {
        
        'Name' : names,
        'Math' : np.random.randint(50, 100),
        'Science' : np.random.randint(50, 100),
        'English' : np.random.randint(50, 100),
    }
    
df['Total'] = df['Math'] + df['Science'] + df['English']

def get_grade(point):
    if point >= 270:
        return "A"
    elif point > 240:
        return "B"
    elif point >= 210:
        return "C"
    elif point >= 180:
        return "D"
    else:
        return "F"

df["Grade"] = df["Total"].apply(get_grade)

df
```
