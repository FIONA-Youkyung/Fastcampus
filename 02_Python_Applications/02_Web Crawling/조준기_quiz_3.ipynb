```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
print(tips.head(), "\n")
print(tips.info(), "\n")
print(tips.describe())
```

```python
# #due to total_bill STD is 8.9, insert a new column of 'bills'
# def bills(total_bill):
#     return (total_bill // 5)

# def tip_apx(tip):
#     return (tip // 0.5)

# tips["bills_approx"] = tips["total_bill"].apply(bills).astype("int")
tips["total_bill"] = tips["total_bill"].round(0).astype("int")
# tips["tip_approx"] = tips["tip"].apply(tip_apx).astype("int")
tips["tip"] = tips["tip"].round(0).astype('int')
tips.head()
```

```python
tips['sex'] = tips['sex'].map({'Female': 0, 'Male': 1})
tips['smoker'] = tips['smoker'].map({'No': 0, 'Yes': 1})
tips['day'] = tips['day'].map({'Thur': 0, 'Fri': 1, 'Sat': 2, 'Sun': 3})
tips['time'] = tips['time'].map({'Lunch': 0, 'Dinner': 1})
tips.head()
```

```python
g = sns.pairplot(tips, hue='tip',  markers='+')
plt.show()
```

```python
g = sns.jointplot(y='tip', x='total_bill', data=tips, kind='reg')
plt.show()
g = sns.violinplot(y='tip', x='total_bill', data=tips, inner = 'quartile')
plt.show()
g = sns.jointplot(y='tip', x='sex', data=tips, kind='reg')
plt.show()
g = sns.violinplot(y='tip', x='sex', data=tips, inner = 'quartile')
plt.show()
g = sns.jointplot(y='tip', x='smoker', data=tips, kind='reg')
plt.show()
g = sns.violinplot(y='tip', x='smoker', data=tips, inner='quartile')
plt.show()
g = sns.jointplot(y='tip', x='day', data=tips, kind='reg')
plt.show()
g = sns.violinplot(y='tip', x='day', data=tips, inner='quartile')
plt.show()
g = sns.jointplot(y='tip', x='time', data=tips, kind='reg')
plt.show()
g = sns.violinplot(y='tip', x='time', data=tips, inner='quartile')
plt.show()
g = sns.jointplot(y='tip', x='size', data=tips, kind='reg')
plt.show()
g = sns.violinplot(y='tip', x='size', data=tips, inner='quartile')
plt.show()
```

# Tips 데이터 시각화 해석

violinplot결과 

 - total_bill은 양의 상관관계, sex는 남자가 조금더 상관관계가 있었고, smoker는 크게 상관이 없었다.
 - day 는 Sat에 tip이 더 많았고, 다른 날들은 큰 영향을 미치지 않았다.
 - time 은 dinner 에 tip이 더 많았다.
 - size는 대체로 tip 이 증가하는 편이지만 3,4 에서 tip이 더 많았다.

heatmap 결과

 - tip과 total_bill, size를 제외하면 correlation관계를 보이기 힘들거나 음의 상관관계를 보였다.
 - total_bill과 tip은 아주 강한 양의 상관관계를 가졌다.
