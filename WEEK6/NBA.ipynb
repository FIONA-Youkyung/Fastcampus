```python
def PremierLeague2(number=20):

    driver = webdriver.Chrome()
    driver.get("https://www.premierleague.com/tables")
    contents = driver.find_elements_by_css_selector("#mainContent .tableBodyContainer tr:not(.expandable)")[:number]
   
    df = pd.DataFrame(columns=["Position", "Club", "Played", "Won", "Drawn", "Lost", "GF", "GA", "GD", "Points"])
    
    for content in contents:
        
        Position = content.find_element_by_css_selector(".value").text 
        Club = content.find_element_by_css_selector(".team .long").text
        Played = content.find_elements_by_css_selector("td")[3].text #td:nth-child(4)
        Won = content.find_elements_by_css_selector("td")[4].text
        Drawn = content.find_elements_by_css_selector("td")[5].text
        Lost = content.find_elements_by_css_selector("td")[6].text
        GF = content.find_elements_by_css_selector("td")[7].text
        GA = content.find_elements_by_css_selector("td")[8].text
        GD = content.find_elements_by_css_selector("td")[9].text
        Points = content.find_elements_by_css_selector("td")[10].text
        
        data = {"Position": Position,
                "Club": Club, 
                "Played": Played,
                "Won": Won, 
                "Drawn": Drawn, 
                "Lost" : Lost,
                "GF" : GF,
                "GA" : GA,
                "GD" : GD,
                "Points" : Points}
        
        df.loc[len(df)] = data
    driver.close()
    return df
```
